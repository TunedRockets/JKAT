Here's some other ways of creating orbits using jkat.

first we need to import the needed packages:

```python
import jkat
%matplotlib qt
# set plotting windows to interactive
```

in addition to creating orbits from the definition, there are several `orbit_from_...` functions that also create orbits.

simplest is `orbit_from_keplerian`, which allows the input of keplerian elements if that is preferred.

```python
ob1 = jkat.orbit_from_keplerian(
    a=10_000,
    e=0.3,
    i=1,
    raan=0,
    argp=0,
    f=2,
    mu=jkat.EARTH_MU,
    t = 5000 # time is optional, but will correctly calculate tp from f (otherwise t=0 is assumed)    
)

ob1
```

There are also two ways of making orbits from external data. the first way is to get a two-line element representation of an orbit.
This represents an orbit around earth, and is available online from several sources.
below is a TLE for the ISS taken from wikipedia:

```python
tle = 'ISS (ZARYA)\n' \
'1 25544U 98067A   08264.51782528 -.00002182  00000-0 -11606-4 0  2927\n' \
'2 25544  51.6416 247.4627 0006703 130.5360 325.0288 15.72125391563537'

ob2 = jkat.orbit_from_tle(tle)
print(ob2)
```

it would be nice to visualize the resulting orbits. fortunately, this is easily doable using `jkat.plot()`. 
this uses pyplot to create 3d visualizations of the orbits. for convenience, the `plt.show()` function is also available as `jkat.show()` so the entire pyplot module doesn't have to be loaded.

```python
jkat.plotting.center(color="blue", s=100)
jkat.plot(ob1, t=0)
jkat.plot(ob2, t=0)
jkat.show()
```

by default, the plot will be rendered from the north pole of the coordinate system. with `jkat.set_view_angle()`, this can be changed.
Several options exist for plotting orbit. the "stilts" of the orbit can be adjusted or removed, and limits can be set so as to only render certain parts

```python
jkat.plotting.center(color="blue", s=100)
jkat.plotting.set_view_angle(45,60)
jkat.plot(ob1, t=0, stilts=False)
jkat.plot(ob2, t=0, f_bounds=(-2,2))
jkat.show()
```
